```python
from transformers import RobertaForTokenClassification
```

```python
#Load model

entity_label_list = ['O', 
         'B-DUR', 'I-DUR', 
         'B-DAT', 'I-DAT', 
         'B-LOC', 'I-LOC', 
         'B-MNY', 'I-MNY', 
         'B-TIM', 'I-TIM', 
         'B-ORG', 'I-ORG', 
         'B-PNT', 'I-PNT', 
         'B-POH', 'I-POH', 
         'B-NOH', 'I-NOH', 
         'B-PER', 'I-PER']
model_path = "/data/bowon_ko/TRoBERTa_BASE/210727/finetune/ner/"
num_labels = len(entity_label_list)
id2label = {str(i): label for i, label in enumerate(entity_label_list)}
label2id = {label:i for i, label in enumerate(entity_label_list)}

model = RobertaForTokenClassification.from_pretrained(
    model_path,
    num_labels= num_labels,
    id2label = id2label,
    label2id = label2id
)
```
