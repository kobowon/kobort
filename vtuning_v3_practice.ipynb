```python
with open('tokenizer/model/wordpiece/version_2.2/vocab.txt', 'r', encoding='utf-8') as f:
    subwords = f.read().split('\n')
print(len(subwords))
```

```python
#이미 적용 완료된 절차
#XSN(명사 접미사), J(조사)는 앞 말과 붙여쓰기 때문에 ♬[X] 라는 subword가 생성될 수 없음
#해당 subword는 오류 문장으로부터 나온 것이라 판단하여 제거함
check = []
for s in subwords[:-1]:
    if s.startswith('♬'):
        check.append(s)
print(check)
```

```python
vocab = {}
for s in subwords[:-1]:
    vocab[s] = 1
```

```python
#duplicate[i][0] 를 처리해도 duplicate[i][1]이 남아있으면 결과적으로 토큰이 분해되지 않음
#duplicate[i][1]가 존재하는 이유가 조사가 분리되지 않아서 일 수도 있고, 명사에 해당 단어가 들어가서 일 수도 있음
#후자의 경우까지 무시한채로  오로지 조사와 접미사 분리에만 초점을 맞춘 vocab-tuning hard 버전을 만든다면 
#아래 방법 적용해보는 것도 방법임
duplicate = []
for i, s in enumerate(subwords[:-1]):
    if '♬' in s:
        tmp = s.split('♬')
        result = ''.join(tmp)
        if result in vocab.keys():
            vocab[result] = 0
            duplicate.append([s, result])

        
```

```python
print(duplicate)
print(len(duplicate))
```
